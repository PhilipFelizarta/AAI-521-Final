```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Model": ["plant_cnn_model.h5", "plant_resnet3_model.h5", "plant_resnet10_model.h5"],
    "Validation Loss": [0.4639, 0.1564, 0.1997],
    "Validation Accuracy (%)": [85.86, 95.06, 93.97],
    "Trainable Parameters": [156070, 912934, 2986278]
}

# Create DataFrame
df = pd.DataFrame(data)

# Define a color palette
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]  # Blue, Orange, Green

# Plot Validation Accuracy
plt.figure(figsize=(10, 6))
plt.bar(df["Model"], df["Validation Accuracy (%)"], color=colors, edgecolor="black", linewidth=1.2)
plt.title("Validation Accuracy of Models", fontsize=18, fontweight='bold')
plt.ylabel("Validation Accuracy (%)", fontsize=14)
plt.xlabel("Model", fontsize=14)
plt.ylim(80, 100)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, acc in enumerate(df["Validation Accuracy (%)"]):
    plt.text(i, acc + 0.5, f"{acc:.2f}%", ha="center", fontsize=12, color="black", weight="bold")
plt.tight_layout()
plt.show()

# Plot Validation Loss
plt.figure(figsize=(10, 6))
plt.bar(df["Model"], df["Validation Loss"], color=colors, edgecolor="black", linewidth=1.2)
plt.title("Validation Loss of Models", fontsize=18, fontweight='bold')
plt.ylabel("Validation Loss", fontsize=14)
plt.xlabel("Model", fontsize=14)
plt.ylim(0, 0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, loss in enumerate(df["Validation Loss"]):
    plt.text(i, loss + 0.01, f"{loss:.4f}", ha="center", fontsize=12, color="black", weight="bold")
plt.tight_layout()
plt.show()

# Plot Trainable Parameters
plt.figure(figsize=(10, 6))
plt.bar(df["Model"], df["Trainable Parameters"], color=colors, edgecolor="black", linewidth=1.2)
plt.title("Trainable Parameters of Models", fontsize=18, fontweight='bold')
plt.ylabel("Trainable Parameters (log scale)", fontsize=14)
plt.xlabel("Model", fontsize=14)
plt.yscale('log')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, params in enumerate(df["Trainable Parameters"]):
    plt.text(i, params * 1.3, f"{params:,}", ha="center", fontsize=12, color="black", weight="bold")
plt.tight_layout()
plt.show()
```
